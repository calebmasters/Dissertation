# Pie chart themes

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('analysisdataset.csv')

# Split the 'Theme_Combined' column into separate themes
df['Theme_Combined'] = df['Theme_Combined'].str.split()

# Create a list of all themes
all_themes = df['Theme_Combined'].explode()

# Map full names to the themes
theme_full_names = {
    'S_int': 'Strong International',
    'S_reg': 'Strong Regional',
    'S_dom': 'Strong Domestic',
    'W_int': 'Weak International',
    'W_reg': 'Weak Regional',
    'W_dom': 'Weak Domestic',
    'AG': 'Green-Growth Critical',
    'NG': 'Non-Growth'
}

# Replace theme abbreviations with full names
all_themes = all_themes.map(theme_full_names)

# Count the occurrences of each theme
theme_counts = all_themes.value_counts()

# Define colors for the pie chart
colors = {
    'Strong International': 'darkblue',
    'Weak International': 'lightblue',
    'Strong Regional': 'purple',
    'Weak Regional': 'pink',
    'Strong Domestic': 'violet',
    'Weak Domestic': 'lightpink',
    'Green-Growth Critical': 'red',
    'Non-Growth': 'orange'
}

# Create a pie chart with the specified order
order = [
    'Strong International', 'Weak International',
    'Strong Regional', 'Strong Domestic',
    'Weak Regional', 'Weak Domestic',
    'Green-Growth Critical', 'Non-Growth'
]

# Sort the theme counts according to the specified order
sorted_theme_counts = theme_counts.reindex(order)

# Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(
    sorted_theme_counts, 
    labels=sorted_theme_counts.index, 
    autopct='%1.1f%%', 
    colors=[colors[key] for key in sorted_theme_counts.index], 
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}, 
    pctdistance=0.85, 
    labeldistance=1.1
)
plt.title('Distribution of Themes in COP Speeches')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('analysisdataset.csv')

# Count the occurrences of each COP event
event_counts = df['Event'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(
    event_counts, 
    labels=event_counts.index, 
    autopct='%1.1f%%', 
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}, 
    pctdistance=0.85, 
    labeldistance=1.1
)
plt.title('Distribution of COP Events')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('analysisdataset.csv')

# Count the occurrences of each continent
continent_counts = df['Region'].value_counts()

# Define colors for each continent
continent_colors = {
    'Africa': 'yellowgreen',
    'Asia': 'gold',
    'Europe': 'lightcoral',
    'North America': 'lightskyblue',
    'South America': 'lightgreen',
    'Oceania': 'orchid',
    'Antarctica': 'lightgrey'
}

# Create a pie chart
plt.figure(figsize=(10, 7))
plt.pie(
    continent_counts, 
    labels=continent_counts.index, 
    autopct='%1.1f%%', 
    colors=[continent_colors.get(continent, 'grey') for continent in continent_counts.index], 
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}, 
    pctdistance=0.85, 
    labeldistance=1.1
)
plt.title('Distribution of Speeches by Continent')
plt.show()
```
