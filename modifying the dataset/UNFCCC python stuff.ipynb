```python
# Checking dirs match

import os

def list_files_by_basename(directory):
    """ Return a set of basenames of all files in a directory, ignoring extensions. """
    files = os.listdir(directory)
    return set(os.path.splitext(file)[0] for file in files)

def compare_directories(dir1, dir2):
    """ Compare two directories to find matching file names, ignoring extensions. """
    files1 = list_files_by_basename(dir1)
    files2 = list_files_by_basename(dir2)
    
    common_files = files1.intersection(files2)
    unique_to_dir1 = files1.difference(files2)
    unique_to_dir2 = files2.difference(files1)

    return common_files, unique_to_dir1, unique_to_dir2

def main():
    directory1 = '/Users/calebmasters/Free From Onedrive mess/Project directory/UNFCCC_speeches/speeches/PDF'
    directory2 = '/Users/calebmasters/Free From Onedrive mess/Project directory/UNFCCC_speeches/speeches/txt'

    common_files, unique_to_dir1, unique_to_dir2 = compare_directories(directory1, directory2)
    
    print("Files with the same name in both directories:", common_files)
    print("Files unique to the first directory:", unique_to_dir1)
    print("Files unique to the second directory:", unique_to_dir2)

if __name__ == "__main__":
    main()
```

# Script to write HTML links and speech data to CSV file

```python
from bs4 import BeautifulSoup
import csv

# The path to the HTML file you provided
html_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/html files/links.html'

# The path where you want the CSV to be saved
output_csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCClinks.csv'

# Function to parse the HTML content and write to CSV
def parse_html_to_csv(html_path, csv_path):
    with open(html_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    # Open the CSV file for writing
    with open(csv_path, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        # Write the headers
        writer.writerow(['Link', 'Language', 'Name of Speech', 'Date'])

        # Iterate over each row in the table body
        for row in soup.select('table tbody tr'):
            # Find the document link and language
            link_tag = row.select_one('a[hreflang]')
            link = link_tag['href'] if link_tag else 'No link available'
            language = link_tag['hreflang'] if link_tag else 'No language specified'

            # Find the name of the speech
            name_of_speech_tag = row.select_one('.views-field-title .documentid')
            name_of_speech = name_of_speech_tag.get_text(strip=True) if name_of_speech_tag else 'No title available'

            # Find the date of the speech
            date_tag = row.select_one('.views-field-field-document-publication-date time')
            date = date_tag['datetime'] if date_tag else 'No date available'

            # Write to CSV
            writer.writerow([link, language, name_of_speech, date])

# Run the function with your file paths
parse_html_to_csv(html_file_path, output_csv_file_path)

print(f"Data has been extracted to {output_csv_file_path}")
```

```python
import pandas as pd
import re

# Load the CSV file
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'  # Replace with the actual path to your CSV file

df = pd.read_csv(csv_file_path)

# Define a function to extract COP number
def extract_cop_number(name):
    # This regex matches 'COP' followed by an optional space and then one or more digits
    match = re.search(r'COP\s?(\d+)', name, re.IGNORECASE)
    # If a match is found, format it as 'COP' followed by the number without any space
    return 'COP' + match.group(1) if match else 'Not a COP event'

# Apply the function to extract and format COP number
df['COP Number'] = df['Name of Speech'].apply(extract_cop_number)

# Overwrite the original CSV file with the modifications
df.to_csv(csv_file_path, index=False)

print(f"The CSV file '{csv_file_path}' has been updated with the COP Number column.")
```

### 2. normalizing the date column

date collumn is currently in the format "2024-03-04T12:00:00Z". The following code normalizes the date to "2024-03-04"

```python
import pandas as pd

# Load the CSV file
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'  # Replace with the actual path to your CSV file

df = pd.read_csv(csv_file_path)

# Custom function to convert date
def convert_date(date_str):
    try:
        # Attempt to convert using the specific format
        return pd.to_datetime(date_str, format="%Y-%m-%dT%H:%M:%SZ", errors='coerce').date()
    except ValueError:
        # Handle cases where conversion fails
        return 'Invalid date'  # Or np.nan, or pd.NaT if you prefer

# Apply the custom function to the 'Date' column
df['Date'] = df['Date'].apply(convert_date)

# Overwrite the original CSV file with the modifications
df.to_csv(csv_file_path, index=False)

print(f"The CSV file '{csv_file_path}' has been updated with the corrected date format.")
```

```python
# detects the correct dates from the pdfs

from pdfminer.high_level import extract_text
from dateutil import parser
import os
import re

# Directory containing the PDFs
pdf_directory = 'UNFCCC speeches/PDF'  # Update this to your PDF directory path

# Output text file path
output_txt_file_path = 'dates.txt'

# Function to extract text from the first page of a PDF
def extract_text_from_first_page(pdf_path):
    try:
        return extract_text(pdf_path, maxpages=1)
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {e}")
        return None

# Function to find a date in a string
def find_date_in_text(text):
    # Regular expression to match a wide range of date formats
    date_patterns = [
        r'\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b',  # Matches dates like 01-01-2020, 1/1/20
        r'\b\d{2,4}[/-]\d{1,2}[/-]\d{1,2}\b',  # Matches dates like 2020-01-01, 20/1/1
        # Add more patterns as needed
    ]
    for pattern in date_patterns:
        match = re.search(pattern, text)
        if match:
            try:
                # Attempt to parse the matched date string into a datetime object
                parsed_date = parser.parse(match.group(), dayfirst=True)  # Adjust dayfirst according to your date format
                # Format the date as dd-mm-yy
                return parsed_date.strftime('%d-%m-%y')
            except ValueError:
                continue
    return 'Date not found'

# Open the output text file for writing
with open(output_txt_file_path, 'w') as output_file:
    # Iterate over all PDF files in the directory
    for filename in os.listdir(pdf_directory):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(pdf_directory, filename)
            text = extract_text_from_first_page(pdf_path)
            if text:
                date = find_date_in_text(text)
                output_file.write(f"{filename}: {date}\n")

print(f"Dates have been extracted and saved to {output_txt_file_path}")
```

### 5. Creating a unique code for each country

```python
import pandas as pd
import re

# Load the CSV file
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'
df = pd.read_csv(csv_file_path)

# Function to sanitize country name (remove spaces and special characters, if necessary)
def sanitize_country_name(country_name):
    if country_name == 'No Country Detected':
        return 'NA'
    else:
        # This example simply replaces spaces with nothing; adjust according to your needs
        return country_name.replace(' ', '')

# Function to extract the unique code from the link
def extract_code_from_link(link):
    # Extract the last 5 or 6 digits from the link
    match = re.search(r'(\d{5,6})$', link)
    return match.group(1) if match else '00000'

def generate_unique_code(row):
    year = 'NA' if row['Date'] == 'Invalid date' else row['Date'].split('-')[0][-2:]
    country_name = sanitize_country_name(row['Detected Country'])
    cop_number = 'NA' if row['COP Number'] == 'Not a COP event' else row['COP Number']
    
    # Extract code from link
    link_code = extract_code_from_link(row['Link'])
    
    return f"{year}{country_name}{cop_number}{link_code}"

df['Unique Code'] = df.apply(generate_unique_code, axis=1)

# Print a small portion of the DataFrame to verify the new column
print(df[['Date', 'Detected Country', 'COP Number', 'Link', 'Unique Code']].head())

# Save the DataFrame to a new CSV file
updated_csv_file_path = 'uniquecodeupdated.csv'
df.to_csv(updated_csv_file_path, index=False)

print("Unique codes generated and saved with modifications.")
```

### 6. Detecting the language of each speech and writing it to the language column

```python
from pdfminer.high_level import extract_text
from langdetect import detect, DetectorFactory, LangDetectException
import pandas as pd
import os

# Ensure consistent results from langdetect
DetectorFactory.seed = 0

def extract_pdf_text(pdf_path):
    try:
        return extract_text(pdf_path)
    except Exception as e:
        return None

def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return None

# Directory containing the PDFs
pdf_directory = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCC speeches/PDF2'

# CSV file path
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'

# Load the CSV file
df = pd.read_csv(csv_file_path)

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    unique_code = row['Unique Code']
    pdf_path = os.path.join(pdf_directory, f"{unique_code}.pdf")
    
    if os.path.exists(pdf_path):
        text = extract_pdf_text(pdf_path)
        if text:
            language = detect_language(text)
            if language:
                df.at[index, 'Language'] = language
    else:
        print(f"PDF {unique_code}.pdf not found.")

# Save the updated DataFrame back to CSV
df.to_csv(csv_file_path, index=False)
```

### Detecting language from the pdf link ###

Language acronym in pdf link:
* SP = "spanish"
* EN = "english"
* ENG = "english"
* AR = "arabic"
* ENJ = "english"
* FR = "french"
* HLE = "english"
* RU = "russian"

```python
import pandas as pd
import re

# Load your DataFrame
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'  # Make sure to update this path
df = pd.read_csv(csv_file_path)

# Define a mapping of acronyms to languages
language_mapping = {
    "SP": "spanish",
    "EN": "english",
    "ENG": "english",
    "AR": "arabic",
    "ENJ": "english",
    "FR": "french",
    "HLE": "english",
    "RU": "russian"
}

# Function to extract the language from the PDF link and update the language column
def update_language(pdf_link):
    # Extract the acronym from the link
    match = re.search(r"HLS_(\w+).pdf$", pdf_link)
    if match:
        # If the acronym is in the mapping, return the corresponding language
        acronym = match.group(1)
        return language_mapping.get(acronym.upper(), "unknown")
    else:
        return "unknown"  # Return "unknown" if no acronym is found or it's not in the mapping

# Update the language column
df['Language'] = df['pdf link'].apply(update_language)

# Save the updated DataFrame back to the CSV
df.to_csv(csv_file_path, index=False)  # Overwrites the original CSV file; change the path if needed

print("Language column updated based on PDF links.")
```

So the pdfs were not renamed at all. the following code renames the pdfs based on the unique code, Ignoring rows in the csv with a broken link.

```python
import pandas as pd
import os
from urllib.parse import unquote  # Import the unquote function

# Load the CSV file, ensuring you update the path
csv_file_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCClinks.csv'
# Specify your download directory
download_directory = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCC speeches/PDF2'

# Load CSV
df = pd.read_csv(csv_file_path)

# Filtering out rows with 'BROKENLINK'
df = df[df['PDF Link'] != 'BROKENLINK']

for _, row in df.iterrows():
    pdf_link = row['PDF Link']
    unique_code = row['Unique Code']
    
    # Use unquote() to decode the URL-encoded filename
    filename_from_url = unquote(pdf_link.split('/')[-1])
    
    # Construct the expected path of the downloaded file
    downloaded_file_path = os.path.join(download_directory, filename_from_url)
    
    # Check if the file exists under the expected name
    if os.path.exists(downloaded_file_path):
        # Extract the file extension to preserve it
        file_extension = os.path.splitext(downloaded_file_path)[1]
        
        # Construct the new file path with the unique code
        new_file_path = os.path.join(download_directory, f"{unique_code}{file_extension}")
        
        # Rename the file
        os.rename(downloaded_file_path, new_file_path)
        print(f"Renamed '{filename_from_url}' to '{unique_code}{file_extension}'")
    else:
        print(f"File '{filename_from_url}' not found in download directory.")

print("All valid files have been renamed.")

```

### Crossreferencing what pdfs in the other directory are unique ###

Need pdfminer6 to be installed. `pip install pdfminer.six`

This code creates a csv file with two columns reperesenting the two direcotories. It contains a list of all the names of the pdfs in both directories and shows overlap between the two.

```python
from pdfminer.high_level import extract_text
import os
import pandas as pd
from concurrent.futures import ProcessPoolExecutor
from concurrent.futures import ThreadPoolExecutor
import concurrent.futures

def extract_pdf_text(pdf_path):
    try:
        print(f"Extracting text from {pdf_path}...")
        text = extract_text(pdf_path)
        print(f"Finished extracting text from {pdf_path}.")
        return pdf_path, text.strip()
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {e}")
        return pdf_path, None


# Then, when you use it:
def get_pdf_texts(directory):
    pdf_texts = {}
    # Collect all PDF paths from subdirectories
    pdf_paths = []
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith('.pdf'):
                pdf_path = os.path.join(root, filename)
                pdf_paths.append(pdf_path)
                
    # Use ThreadPoolExecutor for parallel text extraction
    with ThreadPoolExecutor() as executor:
        future_to_pdf = {executor.submit(extract_pdf_text, pdf_path): pdf_path for pdf_path in pdf_paths}
        for future in concurrent.futures.as_completed(future_to_pdf):
            pdf_path, text = future.result()
            if text:
                # Use the basename of the file path as the key
                filename = os.path.basename(pdf_path)
                pdf_texts[filename] = text
    return pdf_texts


# Directories to compare
dir1 = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCC speeches/PDF2'
dir2 = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/speeches/speeches/Original'

print("Extracting texts from dir1 PDFs...")
dir1_texts = get_pdf_texts(dir1)
print("Extracting texts from dir2 PDFs...")
dir2_texts = get_pdf_texts(dir2)

# Prepare to store matches and non-matches
matches = []

print("Comparing texts between dir1 and dir2...")
for filename1, text1 in dir1_texts.items():
    for filename2, text2 in dir2_texts.items():
        if text1 == text2:
            print(f"Match found: {filename1} and {filename2}")
            matches.append({'Dir1_PDF': filename1, 'Dir2_PDF': filename2})
            break

# Create DataFrame from matches
pdf_matches = pd.DataFrame(matches)

# Save to CSV
csv_output_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/pdfdirecotrycompare.csv'
pdf_matches.to_csv(csv_output_path, index=False)

print(f"CSV file created at {csv_output_path}.")
```

```python
import shutil
import os

import shutil
import os

def move_files_with_name(src_directory, target_directory, file_name_contains):
    """
    Move specified files from a source directory to a target directory.
    
    This function searches the source directory recursively for files whose names exactly match any of the specified names in the file_name_contains list.
    If such files are found, they are moved to the target directory. The target directory is created if it does not exist.
    
    Parameters:
    - src_directory (str): The path to the source directory where the files are located.
    - target_directory (str): The path to the target directory where the files should be moved.
    - file_name_contains (list of str): A list of exact file names. If a file name matches any of these names, it will be moved.
    
    Returns:
    None. Files matching the criteria are moved to the target directory, and a message is printed for each moved file.
    """
    # Ensure the target directory exists
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
    
    # Walk through the source directory
    for root, dirs, files in os.walk(src_directory):
        for file in files:
            # Check if the file name exactly matches any of the specified file names
            if file in file_name_contains:
                # Construct full file path
                file_path = os.path.join(root, file)
                # Construct target file path
                target_path = os.path.join(target_directory, file)
                
                # Move the file
                shutil.move(file_path, target_path)
                print(f"Moved: {file_path} -> {target_path}")

# Example usage
src_directory = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/speeches/speeches/Original'
target_directory = 'Web Scraping/UNFCCC speeches/UNFCCC speeches/PDF2'
file_name_contains = [
    'COP26 United_Arab_Emirates.pdf',
    'COP28 Australia.pdf',
    'CHINESE COP26 China.pdf',
    'ARABIC COP27 Oman.pdf',
    'INCOMPLETE COP28 Austria.pdf',
    'COP25 Australia.pdf',
    'COP26 United_Kingdom.pdf',
    'COP26 Belgium.pdf',
    'COP24 Spain.pdf',
    'COP27 Greece.pdf',
    'COP24 Austria.pdf',
    'COP26 Jordan.pdf',
    'COP26 Uganda.pdf',
    'COP28 Jordan.pdf',
    'FRENCH COP28 Morocco.pdf',
    'ARABIC COP26 Tunisia.pdf',
    'COP28 Armenia.pdf',
    'COP27 Botswana.pdf',
    'COP24 Barbados.pdf',
    'COP27 Bangladesh.pdf',
    'RUSSIAN COP25 Belarus.pdf',
    'COP26 Bahamas.pdf'
    ]

move_files_with_name(src_directory, target_directory, file_name_contains)
```

# Modifying the speech files

Converting to rawtext files

```python
import os
from pdfminer.high_level import extract_text
from docx import Document

def convert_pdf_to_text(pdf_path):
    try:
        return extract_text(pdf_path)
    except Exception as e:
        print(f"Error converting {pdf_path}: {e}")
        return None

def convert_docx_to_text(docx_path):
    try:
        doc = Document(docx_path)
        return "\n".join([paragraph.text for paragraph in doc.paragraphs])
    except Exception as e:
        print(f"Error converting {docx_path}: {e}")
        return None

def save_text_to_file(text, file_path):
    try:
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(text)
    except Exception as e:
        print(f"Error saving {file_path}: {e}")

def convert_files_in_directory(directory_path):
    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        base_name, ext = os.path.splitext(filename)
        text_file_path = os.path.join(directory_path, f"{base_name}.txt")
        
        if ext.lower() == '.pdf':
            text = convert_pdf_to_text(file_path)
            if text:
                save_text_to_file(text, text_file_path)
                print(f"Converted {filename} to text.")
                
        elif ext.lower() == '.docx':
            text = convert_docx_to_text(file_path)
            if text:
                save_text_to_file(text, text_file_path)
                print(f"Converted {filename} to text.")

# Replace 'your_directory_path' with the path to the directory containing your PDF and .docx files
directory_path = '/Users/calebmasters/Library/CloudStorage/OneDrive-Personal/Documents/university/dissertation/coding folder/linguistic analysis/project directory/Web Scraping/UNFCCC speeches/UNFCCC speeches/Rawtext'
convert_files_in_directory(directory_path)
```
